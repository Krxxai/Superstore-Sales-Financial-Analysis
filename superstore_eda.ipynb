```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print("Libraries loaded!")
```

```python
df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')
```

```python
print("Shape:", df.shape)
```

```python
print("\nColumns:", df.columns.tolist())
```

```python
df.head()
```

```python
print("Dataset Info:")
print(df.info())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("\nBasic Stats:")
df[['Sales', 'Profit', 'Discount', 'Quantity']].describe()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Month Name'] = df['Order Date'].dt.strftime('%B')
print("Date columns created!")
df.head()
```

```python
df.info()
```

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
profit_margin = (total_profit / total_sales) * 100
total_orders = df['Order ID'].nunique()
total_customers = df['Customer ID'].nunique()
```

```python
print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Profit Margin: {profit_margin:.2f}%")
print(f"Total Orders:  {total_orders:,}")
print(f"Total Customers: {total_customers:,}")
```

```python
# Group data by Region and calculate key metrics
region_analysis = df.groupby('Region').agg(
    Total_Sales=('Sales', 'sum'),      # Total revenue per region
    Total_Profit=('Profit', 'sum'),    # Total profit per region
    Orders=('Order ID', 'nunique')     # Unique orders per region
).round(2)
```

```python
# Calculate profit margin % for each region
region_analysis['Profit_Margin%'] = (
    (region_analysis['Total_Profit'] / 
     region_analysis['Total_Sales']) * 100
).round(2)
```

```python
# Sort by highest sales first
print("SALES BY REGION:")
print(region_analysis.sort_values('Total_Sales', ascending=False))
```

```python
# Group by product category (Furniture, Technology, Office Supplies)
category = df.groupby('Category').agg(
    Total_Sales=('Sales', 'sum'),      # Revenue per category
    Total_Profit=('Profit', 'sum'),    # Profit per category
    Orders=('Order ID', 'nunique')     # Order count per category
).round(2)
```

```python
# Calculate profit margin per category
category['Profit_Margin%'] = (
    (category['Total_Profit'] / 
     category['Total_Sales']) * 100
).round(2)
```

```python
# Sort by highest sales
print("SALES BY CATEGORY:")
print(category.sort_values('Total_Sales', ascending=False))
```

```python
# Group by sub-category to find granular performance
subcat = df.groupby('Sub-Category').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum')
).round(2)
```

```python
# Calculate profit margin per sub-category
subcat['Profit_Margin%'] = (
    (subcat['Total_Profit'] / 
     subcat['Total_Sales']) * 100
).round(2)
```

```python
# Filter only sub-categories where profit is NEGATIVE
# These are the problem areas losing money for the business
loss_makers = subcat[subcat['Total_Profit'] < 0]

print("LOSS MAKING SUB-CATEGORIES:")
print(loss_makers.sort_values('Total_Profit'))
```

```python
# Also show top 5 most profitable for comparison
print("\nTOP 5 PROFITABLE SUB-CATEGORIES:")
print(subcat.sort_values('Total_Profit', ascending=False).head(5))
```

```python
# Create discount bands to group discount ranges
# This helps us understand how discounts affect profit
df['Discount_Band'] = pd.cut(
    df['Discount'],
    bins=[-0.1, 0, 0.2, 0.4, 1.0],    # Define range boundaries
    labels=[
        'No Discount',       # 0% discount
        'Low (0-20%)',       # 1% to 20% discount
        'Medium (20-40%)',   # 21% to 40% discount
        'High (40%+)'        # Above 40% discount
    ]
)
```

```python
# Calculate average profit and total sales per discount band
discount_impact = df.groupby('Discount_Band').agg(
    Avg_Profit=('Profit', 'mean'),     # Average profit per order
    Total_Sales=('Sales', 'sum'),      # Total revenue
    Orders=('Order ID', 'count')       # Number of orders
).round(2)
```

```python
print("DISCOUNT IMPACT ON PROFIT:")
print("(Higher discount = lower profit?)")
print(discount_impact)
```

```python
# Group by year to see revenue growth over time
yearly = df.groupby('Year').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum')
).round(2)
```

```python
# Calculate year over year growth percentage
# pct_change() automatically calculates % change from previous year
yearly['Growth%'] = (
    yearly['Total_Sales'].pct_change() * 100
).round(2)
```

```python
print("YEARLY REVENUE TREND:")
print(yearly)
```

```python
# Find top 10 customers by total revenue generated
# Important for identifying high value customers
top_customers = df.groupby('Customer Name').agg(
    Total_Sales=('Sales', 'sum'),      # How much they spent
    Total_Profit=('Profit', 'sum'),    # How much profit they generated
    Orders=('Order ID', 'nunique')     # How many times they ordered
).round(2)
```

```python
# Sort and get top 10
top_10 = top_customers.sort_values(
    'Total_Sales', ascending=False
).head(10)

print("TOP 10 CUSTOMERS BY REVENUE:")
print(top_10)
```

```python
# Create a figure with 2 charts side by side
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Chart 1 â Sales by Region (Bar chart)
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
axes[0].bar(region_sales.index, region_sales.values, color=['#1B2A4A','#2563EB','#64748B','#93C5FD'])
axes[0].set_title('Total Sales by Region', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Region')
axes[0].set_ylabel('Total Sales ($)')
# Add value labels on each bar
for i, v in enumerate(region_sales.values):
    axes[0].text(i, v + 1000, f'${v:,.0f}', ha='center', fontsize=9)

# Chart 2 â Profit by Category (Bar chart)
cat_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
colors = ['green' if x > 0 else 'red' for x in cat_profit.values]
axes[1].bar(cat_profit.index, cat_profit.values, color=colors)
axes[1].set_title('Total Profit by Category', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Category')
axes[1].set_ylabel('Total Profit ($)')
# Add value labels
for i, v in enumerate(cat_profit.values):
    axes[1].text(i, v + 500, f'${v:,.0f}', ha='center', fontsize=9)

plt.tight_layout()
plt.savefig('region_category_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved!")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Chart 1 â Average Profit by Discount Band
discount_profit = df.groupby('Discount_Band')['Profit'].mean()
colors = ['green' if x > 0 else 'red' for x in discount_profit.values]
axes[0].bar(discount_profit.index, discount_profit.values, color=colors)
axes[0].set_title('Average Profit by Discount Band', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Discount Band')
axes[0].set_ylabel('Average Profit ($)')
axes[0].axhline(y=0, color='black', linewidth=0.8)  # Zero line
for i, v in enumerate(discount_profit.values):
    axes[0].text(i, v + 0.5, f'${v:,.1f}', ha='center', fontsize=9)

# Chart 2 â Yearly Sales Trend (Line chart)
yearly_sales = df.groupby('Year')['Sales'].sum()
axes[1].plot(yearly_sales.index, yearly_sales.values, 
             marker='o', color='#2563EB', linewidth=2.5, markersize=8)
axes[1].fill_between(yearly_sales.index, yearly_sales.values, alpha=0.1, color='#2563EB')
axes[1].set_title('Yearly Revenue Trend', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Total Sales ($)')
for x, y in zip(yearly_sales.index, yearly_sales.values):
    axes[1].text(x, y + 5000, f'${y:,.0f}', ha='center', fontsize=9)

plt.tight_layout()
plt.savefig('discount_yearly_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved!")
```
